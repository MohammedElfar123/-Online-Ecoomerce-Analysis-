# -Feature Extracting :

```python
# extracting year , month, month name and day from order_date column 

data["year"] = data["order_date"].dt.year

data["month_name"] = data["order_date"].dt.month_name()
data["month"] = data["order_date"].dt.month

data["day"] = data["order_date"].dt.day_name()                              
```

```python
# drop order date 
data.drop("order_date", axis= 1, inplace= True) 
```

```python
data.head()
```

```python
# function to Extract Seasons from month column

def seasons_order(season):
   
 if season in range(1, 4):
    return 'winter'
 elif season in range (4, 7):
    return 'spring'
 elif season in range(7,10):
    return'summer'
 else:
    return'autumn'
 
data['season_order'] = data["month"].apply(seasons_order) 
```

```python
# Function to extract quarter from month 
def quarter(Qrt):
   
 if Qrt in range(1, 4):
    return 'Q1'
 elif Qrt in range (4, 7):
    return 'Q2'
 elif Qrt in range(7,10):
    return'Q3'
 else:
    return'Q4'
 
data['quarter'] = data["month"].apply(quarter) 
```

```python
data.drop("month", axis= 1, inplace= True) 
```

```python
data.head()
```

```python
# Extract Net Profit from total_cost and total_sales

data["net_profit"] = data["total_sales"] - data["total_cost"]  
```
